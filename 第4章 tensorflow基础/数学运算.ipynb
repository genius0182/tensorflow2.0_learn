对于常见的平方和平方根运算，可以使用 tf.square(x)和 tf.sqrt(x)实现。

```python
x = tf.range(5)
x = tf.cast(x,dtype=tf.float32)
x = tf.square(x)
x
```

平方根运算

```python
x = tf.range(5)
x = tf.cast(x,dtype=tf.float32)
tf.sqrt(x)
```
